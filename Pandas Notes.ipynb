```python
df_4.dropna(axis=0,how='any')  #default values ,inplace = true for permanent changes
```

```python
df_4.dropna(axis=1,how='any')  #column dropped which has any NaN value
```

```python
df_4.dropna(axis=1,how='all') #no column has all missing value
```

```python
df_4.dropna(axis=0,how='all',subset=['last']) #any row with missing last value
```

```python
df_4.dropna(axis=0,how='any',subset=['last','age'])
```

```python
df_4.dropna(axis=0,how='all',subset=['last','age']) #both needs to be missing
```

```python
df_4.replace('NA',np.nan,inplace=True)
df_4
```
